### Accuracy

Accuracy is one of the most common evaluation metrics used in classification problems to measure the overall correctness of a model's predictions. It represents the ratio of correctly predicted instances to the total number of instances in the dataset.

> Note that there is a problem with accuracy - it behaves poorly with unbalanced sets. We will generate a set in which most examples will be positive.

````python
#tn - true negative - zmienna r贸wna si 0 i przewidujemy, 偶e jest 0
#fp - false positive - zmienna r贸wna si 0 i przewidujemy, 偶e jest 1
#fn - false negative - zmienna r贸wna si 1 i przewidujemy, 偶e jest 0 (w medycynie ten gorszy scenariusz)
#tp - true positive - zmienna r贸wna si 1 przeiwdujemy, 偶e jest 1
def accuracy(ground_truth, prediction):
    [tn, fp], [fn, tp]= confusion_matrix(ground_truth, prediction)
    return (tp+tn) / (tp+tn+fp+fn)
````

### Precision & recall

Precision and recall are another examples of evaluation metrics that are also commonly used in (binary) classification problems, particularly when dealing with imbalanced datasets or situations where different types of errors have different consequences.
* Precision: it measures the proportion of true positive predictions (correctly predicted positive instances) out of all positive predictions made by the model.* Recall (Sensitivity / true positive rate): it measures the proportion of true positive predictions (correctly predicted positive instances) out of all actual positive instances in the dataset.

````python
# ile spor贸d sklasyfikowanych 1 jest naprawd 1
def precision(ground_truth, prediction):
    [tn, fp], [fn, tp] = confusion_matrix(ground_truth, prediction)
    return tp / (tp+fp)
````

````python
#ile spor贸d prawdziwych 1 zostao sklasyfikowane jako 1
def recall(ground_truth, prediction):
    [tn, fp], [fn, tp] = confusion_matrix(ground_truth, prediction)
    return tp / (tp+fn)
````

### F1-score

To have a comprehensive evaluation, precision and recall are often combined into a single metric called the F1 score, which considers both precision and recall simultaneously. The F1 score provides a balanced assessment by taking the harmonic mean of precision and recall:

F1 Score = 2 * (Precision * Recall) / (Precision + Recall)

The F1 score ranges from 0 to 1, where 1 represents a perfect balance between precision and recall.

````python
def f1score(ground_truth, prediction):
    prec = precision(ground_truth, prediction)
    rec = recall(ground_truth, prediction)

    return 2*prec*rec / (prec+rec)
````

The following function returns a meshgrid, that is, a list of point indices from a given interval, based on the generated dataset.

````python
def get_grid(data):
    x_min, x_max = data[:,0].min()-1, data[:,0].max()+1
    y_min, y_max = data[:,1].min()-1, data[:,1].max()+1
    return np.meshgrid(
        np.arange(x_min, x_max, 0.01),
        np.arange(y_min, y_max, 0.01)
    )
````

````python
def load_dataset(filename, class_column, index_col=None):
  dataset = pd.read_csv(f'ML-datasets/{filename}.csv', index_col=index_col)
  dataset['class'] = dataset[class_column].astype('category').cat.codes
  classes = dataset.pop(class_column).unique()
  return dataset, classes
````

````python
def prepare_dataset(dataset_name):
    params={
        'iris':{'class_column':'species', 'index_col':None},
        'wine':{'class_column':'Class', 'index_col':None},
        'glass':{'class_column':'Type', 'index_col':None},
        'diabetes':{'class_column':'Class', 'index_col':None}
    }

    dataset, classes =load_dataset(dataset_name, **params[dataset_name])
    y =dataset.pop('class')
    X = dataset
    return X, y, classes
````

## Solver & regularization

* **<font color='#ed7d31'>Solver algorithm</font>**: The solver algorithm determines the optimization algorithm used to estimate the coefficients of the logistic regression model - different solvers have different computational properties and convergence behaviors. Typically, the choice of solver depends on the size of the dataset, the number of features, and the computational resources available.
* **<font color='#ed7d31'>Penalty term (L1 or L2)</font>**: Logistic regression can use either L1 regularization (Lasso) or L2 regularization (Ridge). The penalty term affects the loss function and helps control the complexity of the model. Try to play with it esp. if you see that model is overfitted to the training set.
    * L1 regularization encourages sparsity by setting some coefficient values to zero, leading to feature selection and a more interpretable model.
    * L2 regularization shrinks the coefficient values without setting them exactly to zero and is useful when all features are potentially relevant.

## [OPTIONAL] ROC curve and AUC ROC

**<font color='#ed7d31'>ROC (Receiver Operating Characteristic) and AUC (Area Under the Curve)</font>** are commonly used in binary classification to evaluate the performance of a model. ROC is a graphical representation, while AUC is a numerical metric derived from the ROC curve.
* The ROC curve is a plot of the true positive rate (sensitivity) against the false positive rate (1-specificity) at various classification thresholds.
* In the ROC curve, the ideal model would have a curve that hugs the top-left corner, indicating high sensitivity (true positive rate) and low false positive rate.
* AUC (Area Under the Curve): The AUC is a numerical metric that quantifies the overall performance of a model represented by the ROC curve.
    * The AUC value ranges between 0 and 1, where a value of 1 represents a perfect classifier, and a value of 0.5 indicates a random classifier.

````python
def equal_length(column, n_bins):
    """Method to divide a set into subsets of equal frequency"""
    indices, bins = pd.cut(
        column,
        n_bins,
        labels=False,
        retbins=True,
        duplicates='drop'
    )
    return indices.map(dict(enumerate(bins)))

def automatic_binning(column):
    """Division based on histogram"""
    _, bins = np.histogram(column, bins='fd')
    indices = pd.Series(np.digitize(column, bins)-1)
    return indices.map(dict(enumerate(bins)))

def bin_data(X, binning=None):
    if binning is not None:
        X = X.apply(binning)
    return X
````

````python
from functools import total_ordering

def plot_confusion_matrix(
    conf_matrix, classes, title=None, title_appendix='', cmap=plt.cm.Blues
):
    if title_appendix:
      title_appendix = f'({title_appendix})'
    if title is None:
        title = f'Confusion matrix {title_appendix}'

    fig, ax = plt.subplots()
    img = ax.imshow(conf_matrix, cmap=cmap)
    ax.figure.colorbar(img, ax=ax)

    ax.set_xticks(np.arange(conf_matrix.shape[1]))
    ax.set_xticklabels(
        classes, rotation=45, ha='right', rotation_mode='anchor'
    )
    ax.set_yticks(np.arange(conf_matrix.shape[0]))
    ax.set_yticklabels(
        classes, rotation=45, ha='right', rotation_mode='anchor'
    )

    ax.set_title(title)
    ax.set_ylabel('True label')
    ax.set_xlabel('Predicted label')

    fmt='.2f' if conf_matrix.dtype == 'float' else 'd'
    thresh = conf_matrix.max() / 2
    for y, row in enumerate(conf_matrix):
        for x, cell in enumerate(row):
            ax.text(
                x, y, format(cell, fmt),
                ha='center', va='center',
                color='white' if cell > thresh else 'black'
            )
    fig.tight_layout()

    return ax
````

## Prevent overfitting with CV

**<font color='#ed7d31'>Cross-validation</font>** is a resampling technique used in machine learning to assess the performance and generalization ability of a predictive model. It involves partitioning the available data into subsets or folds and repeatedly training and evaluating the model on different combinations of these subsets.

````python
# helper functions

def calculate_metrics(target, prediction, average='macro'):
    accuracy = accuracy_score(target, prediction)
    precision = precision_score(target, prediction, average=average)
    recall = recall_score(target, prediction, average=average)
    f1 = f1_score(target, prediction, average=average)
    mislabeled = (target != prediction).sum()
    total = len(target)
    return accuracy, precision, recall, f1, mislabeled, total

def print_results(metrics, classifier_id='classifier'):
    print(f'Results for {classifier_id}')
    print('----')
    print(f'  Accuracy:  {metrics[0]}')
    print(f'  Precision: {metrics[1]}')
    print(f'  Recall:    {metrics[2]}')
    print(f'  F1 score:  {metrics[3]}')
    print(f'  Mislabeled {metrics[4]} out of {metrics[5]}')
    print('\n')
````

````python
from sklearn.model_selection import KFold, StratifiedKFold

def cross_validate(classifier, kfold, X, y, classes, params=None, talk=True):
    if params is None:
        params = {}
    if classifier.__class__.__name__ == 'ABCMeta':
        clf = classifier(**params)
    else:
        clf = classifier
    predicted = []
    target = []

    for train_index, test_index in kfold.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        clf.fit(X=X_train, y=y_train)
        predicted = np.concatenate((predicted, clf.predict(X_test)))
        target = np.concatenate((target, y_test))

    metrics = calculate_metrics(target, predicted)
    cm = confusion_matrix(target, predicted)
    if talk:
        print_results(metrics, clf.__class__.__name__)
        plot_confusion_matrix(
            cm, classes, title_appendix=clf.__class__.__name__
        )
        plt.show()
    else:
        return metrics, cm
````
